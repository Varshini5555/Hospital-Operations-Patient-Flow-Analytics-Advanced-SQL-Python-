```python
import requests

url = "https://ors.gov.in/patientPortalApi/apidata/dayDashboard"
headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "application/json"
}

response = requests.get(url, headers=headers)
day_data = response.json()

print(day_data)
```

```python
payload = {
    "date": "2026-01-27"
}

response = requests.post(url, json=payload, headers=headers)
```

```python
import pandas as pd
day_data=pd.DataFrame(day_data)
day_data.head(10)
```

```python
from datetime import date
day_data["snapshot_date"] = date.today()
```

```python
day_data.to_excel("day_dashboard_snapshot.xlsx", index=False)
```
